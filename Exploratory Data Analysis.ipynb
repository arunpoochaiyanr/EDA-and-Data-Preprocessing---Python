#### 5.Coversion of column 'published_at' to datetime data type

```python
df["published_at"]= pd.to_datetime(df["published_at"])
```

```python
df["published_at"]
```

#### 6.Creating new column 'published_month' using the data from column 'pulished_at'

```python
df['published_month']=df['published_at'].dt.month
```

```python
df
```

#### 7.Replacing the number into names of months

```python
df['published_month'].replace({1:'Jan', 2:'Feb', 3:'Mar',4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}, inplace = True)
```

```python
df
```

#### 8.Arraging the months in descending order based on total video count

```python
monthly_video_count = df.groupby('published_month')['video_id'].count().sort_values(ascending=False)
print(monthly_video_count)
```

#### 9.Finding the count of unique video_id, channel_id, channel_title

#### 13.Effect of number of views for a video by the people based on Dislikes

```python
corr_view_dislikes = df['view_count'].corr(df['dislikes'])
print(corr_view_dislikes)
```

```python
import matplotlib.pyplot as plot
```

```python
plot.scatter(df['view_count'], df['dislikes'])
plot.xlabel('Views')
plot.ylabel('Dislikes')
plot.title('Correlation between Views and Dislikes')
plot.show()
```

<font size=2 color='blue'> **Inference:** </font>
<font size=2 color='black'> From the corelation coefficient 0.68445 and scatter plot, we could see that the relationship between the number of views and number of dislikes is moderate. And also the relationship is perfectly linear. Also the number views has some effect on the number of dislikes or the the number of views does have some effects on the number of dislikes of the video </font>
